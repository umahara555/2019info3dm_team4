```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
import numpy as np
```

```python
import dataset
X, Y = dataset.load_data()
```

```python
pokemon_labels = np.array([
    "Normal",
    "Fire",
    "Water",
    "Electric",
    "Grass",
    "Ice",
    "Fighting",
    "Poison",
    "Ground",
    "Flying",
    "Psychic",
    "Bug",
    "Rock",
    "Ghost",
    "Dragon",
    "Dark",
    "Steel",
    "Fairy"])
```

```python
y = Y[0]
#Type1しか取ってない
x_train = X[0:700]
y_train = y[0:700]
x_test = X[701:]
y_test = y[701:]
img_list = []
for for_1 in range(18):
    choice_idx = np.random.choice(np.where(y == for_1)[0], 10)
    img_list.append(X[choice_idx])

img_rows, img_cols = 64,64
img_channels = 4
```

```python
for for_1 in range(18):
    fig, ax = plt.subplots(1, 10, figsize=(18, 8))
    for for_2 in range(10):
        ax[for_2].imshow(img_list[for_1][for_2].reshape(img_rows,img_cols, img_channels)) #for_2の値+nでn番目以降のテストデータを出力する．
        ax[for_2].set_title(pokemon_labels[for_1])
        ax[for_2].axis('off')
```

```python
clf = RandomForestClassifier(verbose=True,
                             n_jobs=-1,
                             random_state=2525)
clf.fit(x_train, y_train)
```
