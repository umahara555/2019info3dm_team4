```python
import pandas as pd
from PIL import Image
import numpy as np
import glob
from matplotlib import pylab as plt
from multiprocessing import Pool
import time
```

```python
def _read_image(path):
    img = Image.open(path)
    img = img.resize((64,64))
    return np.array(img)

def _read_images():
    image_paths = glob.glob("/workspace/data/images/*")
    image_paths.sort()

    images_length = len(image_paths)
    images = []
    with Pool() as p:
        for image, i in zip(p.imap(_read_image, image_paths), range(images_length)):
            print(f'\r{i}/{images_length}', end='')
            images.append(image)
        print(f'\r{images_length}/{images_length} done')
    return np.array(images)
```

```python
images = np.array(_read_images())
```

```python
images = np.delete(images, 3, 3)
```

```python

for image,i in zip(images,range(1,len(images)+1)):
    img = Image.fromarray(image)
    img.save(f'/workspace/src/32/images2/{i:04}.png')
```
